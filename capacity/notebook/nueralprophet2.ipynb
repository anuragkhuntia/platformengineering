```python
import pandas as pd
from neuralprophet import NeuralProphet
import plotly.offline as py
```

```python
df = pd.read_csv('../data/datasets/onemonthprocessed.csv')
print(df.head(5))
print(df.tail(5))
print(df.describe())
print(df.dtypes)
print(len(df['y']))
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
model = NeuralProphet()
model.fit(df, freq='s')
```

```python
periods = 30 * 24 * 60 * 60
future = model.make_future_dataframe(df, periods=periods)
```

```python
forecast = model.predict(future)
```
